```python
a = 0.5
ms = 7

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection="mollweide")
    
targRA = coord.Angle(gbt_ra*u.deg)
targDec = coord.Angle(gbt_dec*u.deg)

targRA = targRA.wrap_at(180*u.deg)

ax.plot(targRA.radian, targDec.radian, 'o', label='GBT', zorder=2, markersize=ms, alpha=a)

ax.grid()
ax.set_xticklabels([])
ax.set_xlabel('Right Ascension')
ax.set_ylabel('Declination')
_=ax.legend(loc='upper right')
fig.savefig('mollweide-projection-matches.png', bbox_inches='tight', transparent=False)
```
